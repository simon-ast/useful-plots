```python
import polars as pl
import matplotlib.pyplot as plt
plt.style.use("https://raw.githubusercontent.com/simon-ast/matplotlib-plot-style/main/default_style.mplstyle")

import util as u
```

```python
# SOME GLOBALS FOR CONSISTENT COLOUR
c_sing_stiss = "tab:purple"
c_sing_irac = "tab:red"
c_wakeford_wfc3 = "tab:green"
c_carter_g395h = "tab:orange"
c_carter_prism = "grey"
c_carter_nircam = "goldenrod"
c_carter_niriss = "tab:blue"
c_powell_miri = "k"
```

```python
# READ PRE-JWST FILE DATA
folder = "pre-JWST"
sing_stiss = u.SpectralData(f"{folder}/WASP_39_b_3.11466_3717_2.tbl", reference="Sing et al. (2016)")
sing_irac = u.SpectralData(f"{folder}/WASP_39_b_3.11466_3717_1.tbl", reference="Sing et al. (2016)")
wakeford_wfc3 = u.SpectralData(f"{folder}/WASP_39_b_3.11466_4132_1.tbl", reference="Wakeford et al. (2018)")
```

```python
# READ POST-JWST FILE DATA
folder = "post-JWST"
powell_miri = u.SpectralData(f"{folder}/WASP_39_b_3.11466_5352_3.tbl", reference="Powell et al. (2024)")
carter_niriss_1 = u.SpectralData(f"{folder}/WASP_39_b_3.11466_5502_2.tbl", reference="Carter et al. (2024)")
carter_niriss_2 = u.SpectralData(f"{folder}/WASP_39_b_3.11466_5502_3.tbl", reference="Carter et al. (2024)")
carter_nircam = u.SpectralData(f"{folder}/WASP_39_b_3.11466_5502_1.tbl", reference="Carter et al. (2024)")
carter_nirspec_prism = u.SpectralData(f"{folder}/WASP_39_b_3.11466_5502_6.tbl", reference="Carter et al. (2024)")
carter_nirspec_g395h_1 = u.SpectralData(f"{folder}/WASP_39_b_3.11466_5502_4.tbl", reference="Carter et al. (2024)")
carter_nirspec_g395h_2 = u.SpectralData(f"{folder}/WASP_39_b_3.11466_5502_5.tbl", reference="Carter et al. (2024)")
```

```python
# PLOT THE HST pre-JWST DATA
fig, ax = u.set_figure(size=(10, 6))
u.plot_spectrum(ax, sing_stiss, label="HST STIS (Sing et al. 2016)", c=c_sing_stiss)
u.plot_spectrum(ax, wakeford_wfc3, label="HST WFC3 (Wakeford et al. 2018)", c=c_wakeford_wfc3)
ax.legend()

hst_lims = ax.get_xlim(), ax.get_ylim()

del  fig, ax
```

```python
# PLOT THE SPITZER pre-JWST DATA
fig, ax = u.set_figure(size=(10, 6))
u.plot_spectrum(ax, sing_irac, label="Spitzer IRAC (Sing et al. 2016)", c=c_sing_irac)
ax.legend(loc="upper left")

spitzer_lims = ax.get_xlim(), ax.get_ylim()

del  fig, ax
```

```python
# PLOT EVERYTHING!
fig, ax = u.set_figure(size=(10, 6))
u.plot_spectrum(ax, sing_stiss, label="HST STIS (Sing et al. 2016)", c=c_sing_stiss)
u.plot_spectrum(ax, wakeford_wfc3, label="HST WFC3 (Wakeford et al. 2018)", c=c_wakeford_wfc3)
u.plot_spectrum(ax, sing_irac, label="Spitzer IRAC (Sing et al. 2016)", c=c_sing_irac)
u.plot_spectrum(ax, carter_nircam, label="NIRCam (Carter et al. 2024)", c=c_carter_nircam)
u.plot_spectrum(ax, carter_niriss_1, label="NIRISS (Carter et al. 2024)", c=c_carter_niriss)
u.plot_spectrum(ax, carter_niriss_2, c=c_carter_niriss)
u.plot_spectrum(ax, carter_nirspec_prism, label="NIRSpec PRISM (Carter et al. 2024)", c=c_carter_prism)
u.plot_spectrum(ax, carter_nirspec_g395h_1, label="NIRSpec G395H (Carter et al. 2024)", c=c_carter_g395h)
u.plot_spectrum(ax, carter_nirspec_g395h_2, c=c_carter_g395h)
u.plot_spectrum(ax, powell_miri, label="MIRI LRS (Powell et al. 2024)", c=c_powell_miri)

ax.legend(ncol=2)
del fig, ax
```
